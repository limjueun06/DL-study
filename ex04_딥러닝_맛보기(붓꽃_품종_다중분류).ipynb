### 목표설정
- 아이리스 데이터를 활용해서 다중분류 신경망을 구현해보자
- 다중분류 신경망과 다른 신경망의 구조 차이를 체크해보자

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 데이터 불러오기
from sklearn.datasets import load_iris
```

```python
# 데이터 객체 만들기
iris = load_iris()
iris.keys()
```

```python
# 문제데이터 / 정답데이터 나누기
X = pd.DataFrame(iris['data'], columns = iris['feature_names'])
y = iris['target']
```

```python
# 결측치 확인
X.info()
# 결측치 유무 / 데이터 타입 체크하기
# 결측치 없음 , 전체 데이터 수 150개, 특성 4개
# 데이터타입 -> 실수형 -> 원핫인코딩 진행하지 않아도 된다.
```

```python
# 데이터 분할하기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    random_state=56,
                                                    test_size = 0.2
)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 신경망 설계
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, InputLayer
```

```python
# 뼈대 설정
model = Sequential()
# 입력층
model.add(InputLayer(shape=(4,)))
# 중간층
model.add(Dense(units=32, activation='sigmoid'))
model.add(Dense(units=16, activation='sigmoid'))
model.add(Dense(units=8, activation='sigmoid'))
# 출력층
model.add(Dense(units=3, activation='softmax'))
```

```python
# 모델 컴파일
model.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = 'SGD',
    metrics = ['accuracy']
)
```

```python
# 모델 학습
model.fit(X_train, y_train, epochs=100)
```

```python
# 일반 categorical_crossentropy 사용
# - 정답데이터에 원핫인코딩 필수
# Tensorflow에서 지원하는 원핫인코딩 함수 사용해보기
from tensorflow.keras.utils import to_categorical
```

```python
y_train_oh = to_categorical(y_train)
y_test_oh = to_categorical(y_test)
```

```python
y_train_oh
```

```python
y_test_oh
```

```python
# 뼈대 설정
model = Sequential()
# 입력층
model.add(InputLayer(shape=(4,)))
# 중간층
model.add(Dense(units=32, activation='sigmoid'))
model.add(Dense(units=16, activation='sigmoid'))
model.add(Dense(units=8, activation='sigmoid'))
# 출력층
model.add(Dense(units=3, activation='softmax'))
```

```python
# 모델 컴파일
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'SGD',
    metrics = ['accuracy']
)
```

```python
# 모델 학습
model.fit(X_train, y_train_oh, epochs=100)
```

```python
# 현재 모델의 성능이 안좋다.
# 오후에 배울 내용을 미리 코드로 적용해서 성능을 끌어올려보자.

# 뼈대 설정
model = Sequential()
# 입력층
model.add(InputLayer(shape=(4,)))
# 중간층
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=8, activation='relu'))
# 출력층
model.add(Dense(units=3, activation='softmax'))
```

```python
# 모델 컴파일
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
# 모델 학습
model.fit(X_train, y_train_oh, epochs=100)
```
