### 문제 3-1. 다음과 같은 조건에 맞는 데이터를 학습하는 딥러닝 모델을 설계하시오.
1. 특성데이터는 특성이 10개로 구성되어있다.
2. 라벨데이터는 클래스가 3개로 원핫인코딩 되어있다.
3. 입력층, 은닉층, 출력층으로 신경망을 설계한다. (출력의 수는 임의로 설정)
4. 입력층, 은닉층의 활성화 함수는 relu를 사용한다.

```python
# 1. 모델 설계
# 뼈대 생성
model = Sequential()

# 입력층 쌓기
model.add(InputLayer(shape=(10,)))

# 은닉층 쌓기
model.add(Dense(units=10, activation='relu'))

# 출력층 쌓기
model.add(Dense(units = 3, activation='softmax'))
```

### 문제 3-2. 설계한 모델을 컴파일하고 학습을 수행하는 코드를 작성하시오.
1. 경사하강법은 adam을 사용한다.
2. 반복수는 100회, batch 크기는 20으로 설정한다.
3. 검증데이터로는 원 데이터의 25%를 사용한다.

```python
# 2-1. 모델 컴파일
model.compile(
    loss='categorical_crossentropy',
    optimizer = 'adam',
    metrics=['accuracy']
)

# 2-2. 모델학습
model.fit(X_train, y_train, validation_split=0.25, epochs=100, batch_size=20)
```

### 문제 3-3. 학습한 모델의 정확도와 예측값을 출력하는 코드를 작성하시오.

```python
# 3-1. 모델 예측
pred = model.predict(X_test)
print(pred)

# 3-2. 모델 정확도 확인
loss, accuracy = model.evaluate(X_test, y_test)
print(loss)
print(accuracy)
```
